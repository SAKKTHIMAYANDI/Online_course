```python
#6

b = []
z = 0
while True:
    a = float(input('Enter the letter:'))
    if a == 0:
         break
    b.append(a)
    z += a
average = z / len(b)
c = [] ## below
d = [] ## equal
e = [] ## above
for i in b:
    if i < average:
        c.append(i)
    elif i == average:
        d.append(i)
    else:
        e.append(i)
    
    
print('below values:',c)
print('average values:',d)
print('above values:',e)
```
