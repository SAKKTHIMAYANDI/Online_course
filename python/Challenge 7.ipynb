```python
#1
class number:
    def set_data(self,v):
        self._value = v
    def display_data(self):
        if self._value > 0:
            print('Its integer value: ',self._value)
        else:
            print('Its negative value: ',self._value)
        return self._value
```

```python
n = number()
n.set_data(int(input('Enter the number: ')))

n.display_data()
```

```python
#2
import math
class calculate:
    def set_data(self,r):
        self._radius = r
    def display_data(self):
        area = 3.14*self._radius**2
        print('Display the the area of perimeter: ',area)
        circum = 2*3.14*self._radius
        print('Display the area of circumference: ',circum)
        Triangle = math.sqrt(3)/4*self._radius**2
        print('Display the regular shapes in triangle: ',Triangle)
```

```python
c1 = calculate()
c1.set_data(int(input('Enter the radius value: ')))

c1.display_data()
```

```python
#3
class student:
    def data_members(self,r,n,b):
        self._rollno = r
        self._sname  = n
        self._branch = b
    def display_data(self):
        print('Roll_no = ',self._rollno)
        print('Student_Name = ',self._sname)
        print('Branch = ',self._branch)
    def __del__(self):
        print('Deleting the object ',+str(self))
```

```python
s1 = student()
s1.data_members(201,' Mayan ','MCA')

s1.display_data()
print('___'*25)

s2 = student()
s2.data_members(202,' Balan','B.E')

s2.display_data()
```

```python
#4
class student1:
    
    count = 0
    
    def data_members(self,r,n,b):
        student1.increament_count()
        self._rollno = r
        self._sname  = n
        self._branch = b
    def display_data(self):
        print('Roll_no = ',self._rollno)
        print('Student_Name = ',self._sname)
        print('Branch = ',self._branch)
    def __del__(self):
        print('Deleting the object ',+str(self))
    @classmethod
    def increament_count(cls):
        cls.count += 1
```

```python
#6
class Date:
    def set_date(self,d = 0,m = 0,y = 0):
        self._days  = d
        self._months = m
        self._years  = y
    def data_memebers(self,d,m,y):
        self._day   = d
        self._month = m
        self._year  = y
    def display_date(self):
        date  = (self._days,self._months,self._years)
        dates = (self._day,self._month,self._year)
        
        if date == dates:
            print(date,'It is Equal',dates)
        else:
            print(date,'It is not  Equal',dates)
```

```python
d1 = Date()
d1.set_date()
d1.set_date(22,2,2001)
d1.data_memebers(23,2,2001)

d1.display_date()
```

```python
#7
class weather:
    def set_weather(self,w = 'snow'):
        self._weather = w
    def display_weather(self):
        if 'snow' == self._weather :
            print('It is snow weather')
        elif 'winter' == self._weather:
            print('It is winter weather')
        elif 'summer' == self._weather:
            print('It is summer weather')
        elif 'rain' == self._weather:
            print('It is rain weather')
        else:
            print('It is not weather')
```

```python
w1 = weather()
w1.set_weather()

w1.display_weather()
```

```python
#8
class complex:
    def __init__(self,r='2+4',i='2+4j'):
        self._real = r
        self._imaginary = i
    def __eq__(self):
        return self._real == self._imaginary
```

```python
c1 = complex()
c1.__init__()

c1.__eq__()
```

```python
#9
class department:
    def set_dept(self):
        self._deptid = input('Enter the department_id: ')
        self._deptname = input('Enter the department_name: ')
    def display_dept(self):
        print('department_id =  ',self._deptid)
        print('department_name = ',self._deptname)
class student1:
    def data_members(self):
        self._rollno = input('Enter the Roll_no: ')
        self._sname  = input('Enter the Student_name: ')
        self._branch = input('Enter the Branch: ')
        self.stu = department()
        self.stu.set_dept()
    def display_data(self):
        print('Roll_no = ',self._rollno)
        print('Student_Name = ',self._sname)
        print('Branch = ',self._branch)
        self.stu.display_dept()
    def __del__(self):
        print('Deleting the object ',+str(self))
```

```python
s1 = student1()
s1.data_members()
print('__'*25)

s1.display_data()
```

```python
#10

class person:
    def __init__(self,n,a):
        self._name = n
        self._age = a
    def birthday(self):
        print('Happie birthday you were: ',self._age)
        self._age += 1
        print('now there person age is: ',self._age)
    def __str__(self):
        return  self._name +' is '+str(self._age)
class student(person):
    def __init__(self,n,a,i):
        super().__init__(n,a)
        self._id = i
    def cal_pay(self,hrs_wrks):
        return hrs_wrks * 2.02
```
