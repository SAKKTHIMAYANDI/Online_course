```python
#3
def medianfun(num1,num2,num3):
    if num1 <= num2 <= num3 or num3 <= num2 <= num1:
        return num2
    elif num2 <= num1 <= num3 or num3 <= num1 <= num2:
        return num1
    else:
        return num3
def main():
    num1 = int(input('Enter the numbers num1 : '))
    num2 = int(input('Enter the numbers num2 : '))
    num3 = int(input('Enter the numbers num3 : '))
    
    median = medianfun(num1,num2,num3)
    
    print('median = ',median)
main()
```

```python
#4
def length(a,b,c):
    if a +b>=c and  b+c >= a and c+a >= b:
        return True
    else:
        return False
a = int(input('Enter the numbers a: '))
b = int(input('Enter the numbers b: '))
c = int(input('Enter the numbers c: '))
if length(a,b,c):
    print('Triangle is valid')
else:
    print('Triangle is invalid')
```

```python
#5
def isinteger(s):
    s=s.strip()
    
    if len(s) < 1:
        return False
    if s[0] == '+' or s[0] == '-':
        s = s[1:]
    if len(s) < 1:
        return False
    return s.isdigit()
def main():
    a = input('Enter the value: ')
    if isinteger(a):
        print('string represented integer')
    else:
        print('string represented not integer')
main()
```

```python
#6
def isprime(n):
    p = True
    for i in range(2,int(n**0.5)+1):
        if n % i == 0:
            p = False
            break
    return p
def nextprime(n):
    a = n + 1
    while not isprime(a):
        a += 1
    return a
def main():
    b = int(input('Enter the numbers: '))
    prime = nextprime(b)
    print('The first prime number larger then',b,'is',prime)
    
main()
        
```

```python
#7

def good_pass(words):
    if len(words) < 8:
        return False
    upper_letter = False
    lower_letter = False
    numbers = False
    
    for char in words:
        if char.isupper():
            upper_letter = True
        elif char.islower():
            lower_letter = True
        elif char.isdigit():
            numbers = True
            
            return upper_letter and lower_letter  and numbers
            
def main():
    user = input('Enter the password: ')
    if good_pass(user):
        print('Its create succesfully')
    else:
        print('recreate a password')
        
main()
```

```python
#10

def a(b):
    if len(b) == 0:
        return ''
    elif len(b) == 1:
        return b[0]
    elif len(b) == 2:
        return b[0] + ' and ' +b[1]
    else:
        c = ','.join(b[:-1])
        return c +', and ' + b[-1]
def main():
    b=[]
    while True:
        n = input('Enter the values: ')
        if n == '':
            break
        b.append(n)
        result = a(b)
        print('Result:',result)
        
main()
```
