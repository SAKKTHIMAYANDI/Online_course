```python
#7 count a month days;
# 31 days months -> jan,mar,may,jul,aug,oct,dec
# 30 days months -> apr,jun,sep,nov
# feb is either 28 or 29 days

month = input('the number of days in that month : ')

if month == 'January' or  month == 'March' or month == 'May' or  month == 'July' or month == 'August'or month == 'October' or  month == 'December':
    print('31 days',month)
elif month == 'April' or month == 'June' or month == 'September' or  month == 'November':
    print('30 days',month)
elif month == 'February':
    print('28 or 29 days',month)
else:
    print('This month is not valid')
```
