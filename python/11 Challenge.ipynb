```python
#4
w=Tk()
def rc_a():
    area = (float(t1.get(1.0,'end'))**2)*3.14
    l.config(text=' Area of circle '+str(area))
    t1.delete(1.0,'end')
l1 =Label(w,text = 'Enter the Radius ')

l1.grid(row=0,column=1)

t1 = Text(w,height=1,width=20)

t1.grid(row=0,column=2)

b1 = Button(w,text='Convert',command = rc_a)
b1.grid(row=2,column=1)

l = Label(w)

l.grid(row = 2, column=0)
w.mainloop()
```

```python
#5 

w = Tk()
def temp():
    F = (float(t1.get(1.0,'end'))*9/5)+32
    l.config(text='temperature in Fahrenheit'+str(F))
    t1.delete(1.6,'end')
    
l1 = Label(w,text='Enter the celsius ')
l1.grid(row=0,column=0)
t1 = Text(w,height=1,width=20)
t1.grid(row=0,column=1)
b1=Button(w,text = 'convert',command=temp)
b1.grid(row=1,column=1)

l = Label(w)
l.grid(row=0,column=2)
w.mainloop()
```

```python
#6

w = Tk()
def kilo():
    F =  float(t1.get(1.0,'end')) / 1.6
    l.config(text='distance in miles'+str(F))
    t1.delete(1.6,'end')
    
l1 = Label(w,text='Enter the KM ')
l1.grid(row=0,column=0)
t1 = Text(w,height=1,width=20)
t1.grid(row=0,column=1)
b1=Button(w,text = 'convert',command=kilo)
b1.grid(row=2,column=1)

l = Label(w)
l.grid(row=2,column=0)
w.mainloop()
```

```python
#7

w = Tk()
def hello():
    print('Welcome')
def  save():
    print('Saved')
def  go():
    print('Exited')
menu = Menu(w)
filemenu = Menu(menu)
filemenu.add_command(label = 'open',command=hello)
filemenu.add_separator()
filemenu.add_command(label = 'save',command=save)
filemenu.add_separator()
filemenu.add_command(label = 'Exit',command=go)
filemenu.add_separator()
menu.add_cascade(label ='file',menu = filemenu )
w.config(menu = menu)
w.mainloop()
```
