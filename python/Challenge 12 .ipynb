```python
#2
def table():
    con = sqlite3.connect('temp_db.db')
    cur = con.cursor()
    cur.execute('create table if not exists emp(empno integer,ename text,salary decimal,deptno integer)')
    con.commit()
```

```python
table()
```

```python
#3
def insert(n,a,s,d):
    con = sqlite3.connect('temp_db.db')
    cur = con.cursor()
    cur.execute('insert into emp values(?,?,?,?)',(n,a,s,d))
    con.commit()
    con.close()
```

```python
insert(101,'Siva',24500,1)
insert(102,'Mayan',15500,2)
insert(103,'John',14500,1)
insert(104,'Sri',26500,2)
insert(105,'Kumar',26000,2)
```

```python
#4
def view():
    con = sqlite3.connect('temp_db.db')
    cur = con.cursor()
    cur.execute('select * from emp')
    row = cur.fetchall()
    con.commit()
    con.close()
    return row
```

```python
view()
```

```python
#5
def view1(s,d):
    con = sqlite3.connect('temp_db.db')
    cur = con.cursor()
    cur.execute('select * from emp where salary > ? and deptno = ?',(s,d))
    row = cur.fetchall()
    con.commit()
    con.close()
    return row
```

```python
view1(15000,10)
```

```python
#6
def view2(s,sal,d):
    con = sqlite3.connect('temp_db.db')
    cur = con.cursor()
    cur.execute('select * from emp where salary >= ? and salary <= ? and deptno != ?',(s,sal,d,))
    row = cur.fetchall()
    con.commit()
    con.close()
    return row
```

```python
view2(20000,25000,10)
```

```python
#9
def update(n1,n2):
    con = sqlite3.connect('temp_db.db')
    cur = con.cursor()
    cur.execute('update emp set salary = salary * ? where empno = ?',(n1,n2))
    row = cur.fetchall()
    con.commit()
    con.close()
    return row
```
