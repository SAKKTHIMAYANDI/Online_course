```python
#3
class Positive:
    def __init__(self,l):
        self._limit = l
        self._value = 1
    def __iter__(self):
        return self
    def __next__(self):
        if self._value > self._limit:
            raise StopIteration
        else:
            return_value = self._value * self._value
            self._value += 1
            return return_value
```

```python
n = int(input('Enter the integer value: '))
try:
    if n < 0:
        raise ValueError(' Enter the value above 0 ')
except ValueError as e:
    print(e)
for i in Positive(n):
    print(i,end = ' ')
```

```python
#5
class Iter_number:
    def __init__(self,l):
        self._limit = l
        self._value = 1
    def __iter__(self):
        return self
    def __next__(self):
        if self._value > self._limit:
            raise StopIteration
        else:
            return_val = self._value
            self._value += 1
            return return_val
```

```python
n = int(input('Enter the integer values: '))

for i in Iter_number(n):
    print(i,end = ',')
```

```python
#6
class Prime:
    def __init__(self,l):
        self._limit = l
        self._value = 2
    def __iter__(self):
        return self
    def prime(self,num):
        for i in range(2,num):
            if (num%i) == 0:
                break
        else:
            return num
    def __next__(self):
        if self._value > self._limit:
            raise StopIteration
        else:
            return_val = self.prime(self._value)
            self._value += 1
            return return_val
```

```python
n = int(input('Enter the integer value: '))

for i in Prime(n):
    if i != None:
        print(i)
```

```python
#7
class Circle:
    def __init__(self,s,n):
        self._sequence = s
        self._number = n
        self._index = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self._index >= self._number:
            raise StopIteration
        else:
            return_value = self._sequence[self._index % len(self._sequence)]
            self._index += 1
            return return_value
```

```python
c1 = Circle('sakkthi',5)
print(list(c1))
```

```python
#8
class Maya:
    def __init__(self,n):
        self._num = n
        self._value = 10
    def __iter__(self):
        return self
    def __next__(self):
        if self._value > self._num:
            raise StopIteration
        else:
            return_val = self._value
            self._value += 1
            return return_val
```

```python
n = int(input('Enter the integer number: '))

for i in Maya(n):
    print(i,end = ' ')
```

```python
#9
def Start(fun):
    def value(*args,**kwargs):
        print('**'*25)
        fun(*args,**kwargs)
        print('**'*25)
    return value
def End(func):
    def values(*args,**kwargs):
        print('__'*25)
        func(*args,**kwargs)
        print('__'*25)
    return values
```

```python

@Start
@End
def display(msg):
    print(msg)
display('sakthi')
```

```python
#10
import time
def timer(sets):
    def calculates(*args,**kwargs):
        Start_time = time.perf_counter()
        check = sets(*args,**kwargs)
        End_time = time.perf_counter()
        runtime = End_time - Start_time
        print('Execution Time :::{0:20.10f}secs'.format(runtime))
        return check
    return calculates
```

```python
@timer
def display(check):
    print(check)
display('Time')
```
