```python
#13

import random

choice = ['rock', 'paper', 'scissors']
user_score = 0
computer_score = 0

while user_score < 3 and computer_score < 3:
    
    play_choice = input("Enter your choice: ").lower()

    if play_choice not in choice:
        print('invalid.please try again')
        continue
        
    computer_choice = random.choice(choice)
    print(f'your choice{play_choice}')
    print(f'computer {computer_choice}')
          
          
    if play_choice == computer_choice:
        if computer_choice == 'Paper':
            print('it\'s a tie.')
    elif (play_choice == 'rock' and computer_choice == 'scissors') or (play_choice == 'paper' and computer_choice == 'rock') or (play_choice == 'scissors' and computer_choice == 'paper'):
          print(' You wins the game')
          user_score += 1
    else:
          print('Computer win the game')
          computer_score += 1
          
    print(f'your score: {user_score}')
    print(f'computer score: {computer_score}')

          
if user_score == 3:
    print(f'you win {user_score}')
else:
    print(f'computer win {computer_score}')
```
